```python
from tkinter import *
from PIL import Image, ImageTk
from tkinter import messagebox
from tkinter.ttk import Combobox
```

```python
win = Tk()
win.state('zoomed')
win.resizable(width=False,height=False)
header_frm=Frame(win)
header_frm.place(x=0,y=0,relwidth=1,relheight=.2)

img_bg = Image.open('d:/ducat.jpg')
tk_imgbg = ImageTk.PhotoImage(img_bg.resize((1370,145)))
lbl_bg=Label(header_frm,image=tk_imgbg)
lbl_bg.pack()

# 1st page which will get the username and password.

def main_body():
    def reset():
        user_entry.delete(0,"end")
        pass_entry.delete(0,"end")
        user_entry.focus()
        
    def login():
        u=user_entry.get()
        p=pass_entry.get()
        if(len(u)==0 or len(p)==0):
            messagebox.showwarning('validation',"Username/Password can't be empty")
        else:
            if(u=='admin' and p=='admin'):
                frm.destroy()
                login_body()
            else:
                messagebox.showerror('validation',"Invalid Username/Password")
    
    frm = Frame(win)
    frm.place(x=0,rely=.2,relwidth=1,relheight=.8)


    img_lp = Image.open('d:/design.jpg')
    tk_imglp = ImageTk.PhotoImage(img_lp.resize((1400,1000)))
    lbl_lp = Label(frm,image=tk_imglp)
    lbl_lp.image=tk_imglp
    lbl_lp.pack()

    frm1 = Frame(frm)
    frm1.configure(bg='yellow')
    frm1.place(relx=.4,rely=.2,relwidth=.15,relheight=.3)

    user_lbl=Label(frm,text="Username:",font=('',12,'bold'),bg='yellow',fg='black')
    user_lbl.place(relx=.42,rely=.22)

    user_entry=Entry(frm,font=('','10'),bd=3)
    user_entry.focus()
    user_entry.place(relx=.42,rely=.26)

    pass_lbl=Label(frm,text="Password:",font=('',12,'bold'),bg='yellow',fg='black')
    pass_lbl.place(relx=.42,rely=.32)

    pass_entry=Entry(frm,font=('','10'),bd=3,show='*')
    pass_entry.place(relx=.42,rely=.36)

    login_btn=Button(frm,text="Login",font=('','10'),bd=3,command=login)
    login_btn.place(relx=.43,rely=.41)

    reset_btn=Button(frm,text="Reset",font=('','10'),bd=3,command=reset)
    reset_btn.place(relx=.48,rely=.41)

# 2nd page which will show logout, register, find, deposit and add button

def login_body():
    def logout():
        frm.destroy()
        main_body()
        
    def register_student():
        frm.destroy()
        register_student_body()
        
    def find():
        frm.destroy()
        find_student_body()
    
    def deposit():
        frm.destroy()
        deposit_fee_body()
    
    def add():
        frm.destroy()
        add_course_body()
        
        
    frm=Frame(win)
    frm.place(x=0,y=0,relwidth=1,relheight=1)
    
    img_lp = Image.open('d:/entry-4.png')
    tk_imglp = ImageTk.PhotoImage(img_lp.resize((1375,850)))
    lbl_lp = Label(frm,image=tk_imglp)
    lbl_lp.image=tk_imglp
    lbl_lp.pack()

    wel_lbl=Label(frm,text="Welcome, Admin",font=('',12),bg='black',fg='white')
    wel_lbl.place(relx=0,rely=0)
    
    logout_btn=Button(frm,text="Logout",font=('',10),bd=3,command=logout)
    logout_btn.place(relx=.96,rely=0)
    
    reg_btn=Button(frm,text="Register Student",font=('',12),bd=3,width=15,command=register_student)
    reg_btn.place(relx=.7,rely=.1)
    
    find_btn=Button(frm,text="Find Student",font=('',12),bd=3,width=15,command=find)
    find_btn.place(relx=.7,rely=.3)
    
    dep_btn=Button(frm,text="Deposit Fee",font=('',12),bd=3,width=15,command=deposit)
    dep_btn.place(relx=.7,rely=.5)
    
    course_btn=Button(frm,text="Add Course",font=('',12),bd=3,width=15,command=add)
    course_btn.place(relx=.7,rely=.7)

# page 3 where we need student info
    
def register_student_body():
    def logout():
        frm.destroy()
        main_body()
    
    def home():
        frm.destroy()
        login_body()
        
def register_student_body():
    def logout():
        frm.destroy()
        main_body()
    def home():
        frm.destroy()
        login_body()
        
    def register_stu_db():
        stu_id=id_entry.get()
        stu_name=name_entry.get()
        stu_mob=mob_entry.get()
        stu_crs=course_list.get()
        stu_fee=fee_entry.get()
        crs,crs_fee=stu_crs.split()
        con=sql.connect(database="studentdb.sqlite")
        cursor=con.cursor()
        cursor.execute("insert into students values(?,?,?,?,?,?)",(stu_id,stu_name,stu_mob,crs,crs_fee,(int(crs_fee)-int(stu_fee))))
        con.commit()
        con.close()
        messagebox.showinfo('Registration',"Student Registered...")
        frm.destroy()
        register_student_body()
       
    frm=Frame(win)
    frm.place(x=0,y=0,relwidth=1,relheight=1)

    img_lp = Image.open('d:/reg-2.jpg')
    tk_imglp = ImageTk.PhotoImage(img_lp.resize((1375,850)))
    lbl_lp = Label(frm,image=tk_imglp)
    lbl_lp.image=tk_imglp
    lbl_lp.pack()
    
    logout_btn=Button(frm,text="Logout",font=('',10),bd=3,command=logout)
    logout_btn.place(relx=.96,rely=0)
    
    home_btn=Button(frm,text="Home",font=('',10),bd=3,command=home)
    home_btn.place(x=0,y=0)
    
    con=sql.connect(database="studentdb.sqlite")
    cursor=con.cursor()
    cursor.execute("select max(stu_id) from students")
    max_id=cursor.fetchone()
    if(max_id[0]==None):
        stu_id_db=1
    else:
        stu_id_db=max_id[0]+1
    con.close()
    stu_id=Label(frm,text="Student Id:",font=('',12,'bold'),bg='black',fg='white')
    stu_id.place(relx=.3,rely=.2)
    
    id_entry=Entry(frm,font=('',12),bd=3)
    id_entry.place(relx=.47,rely=.2)
    id_entry.insert(0,str(stu_id_db))
    id_entry.configure(state='disabled')
    
    stu_name=Label(frm,text="Student Name:",font=('',12,'bold'),bg='black',fg='white')
    stu_name.place(relx=.3,rely=.3)
    
    name_entry=Entry(frm,font=('',12),bd=3)
    name_entry.focus()
    name_entry.place(relx=.47,rely=.3)
    
    stu_mob=Label(frm,text="Student Mob:",font=('',12,'bold'),bg='black',fg='white')
    stu_mob.place(relx=.3,rely=.4)
    
    mob_entry=Entry(frm,font=('',12),bd=3)
    mob_entry.place(relx=.47,rely=.4)
    
    stu_course=Label(frm,text="Select Course:",font=('',12,'bold'),bg='black',fg='white')
    stu_course.place(relx=.3,rely=.5)
    
    con=sql.connect(database="studentdb.sqlite")
    cursor=con.cursor()
    cursor.execute("select * from course")
    courses=cursor.fetchall()
    courses.insert(0,"---select----")
    con.close()
    course_list=Combobox(frm,values=courses,font=('',12,'bold'))
    course_list.place(relx=.47,rely=.5)
    course_list.current(0)
    
    reg_fee=Label(frm,text="Reg. Amt:",font=('',12,'bold'),bg='black',fg='white')
    reg_fee.place(relx=.3,rely=.6)
    
    fee_entry=Entry(frm,font=('',12),bd=3)
    fee_entry.place(relx=.47,rely=.6)
    
    reg_btn=Button(frm,text="Register",font=('',15),bg='gray',fg='black',bd=3,command=register_stu_db)
    reg_btn.place(relx=.4,rely=.75)

    
# Find Student

def find_student_body():
    def logout():
        frm.destroy()
        main_body()
    def home():
        frm.destroy()
        login_body()
    
    def search():
        stu_id=id_entry.get()
        con=sql.connect(database="studentdb.sqlite")
        cursor=con.cursor()
        cursor.execute("select * from students where stu_id=?",(stu_id,))
        row=cursor.fetchone()
        if(row==None):
            messagebox.showwarning("Result","Student does not exist")
        else:
            messagebox.showinfo("Result",f'Student name : {row[1]}\nStudent Mobile no.: {row[2]}\nStudent Course:{row[3]}\nStudent course fee: {row[4]}\nStudent Remaining Fee: {row[5]}')
            
    frm=Frame(win)
    frm.place(x=0,y=0,relwidth=1,relheight=1)
    
    img_lp = Image.open('d:/find.jpg')
    tk_imglp = ImageTk.PhotoImage(img_lp.resize((1375,850)))
    lbl_lp = Label(frm,image=tk_imglp)
    lbl_lp.image=tk_imglp
    lbl_lp.pack()
    
    logout_btn=Button(frm,text="Logout",font=('',10),bd=3,command=logout)
    logout_btn.place(relx=.96,rely=0)
    
    home_btn=Button(frm,text="Home",font=('',10),bd=3,command=home)
    home_btn.place(x=0,y=0)
    
    frm1 = Frame(frm)
    frm1.configure(bg='powder blue')
    frm1.place(relx=.6,rely=.37,relwidth=.3,relheight=.28)
    
    stu_id=Label(frm,text="Student id:",font=('',12,'bold'),bg='powder blue')
    stu_id.place(relx=.65,rely=.45)
    
    id_entry=Entry(frm,font=('',12),bd=3)
    id_entry.focus()
    id_entry.place(relx=.73,rely=.45)
    
    search_btn=Button(frm,text="Search",font=('',13),bd=3,command=search)
    search_btn.place(relx=.70,rely=.52)

# fee deposit

def deposit_fee_body():
    def logout():
        frm.destroy()
        main_body()
    def home():
        frm.destroy()
        login_body()
        
    def fetch_fee(event):
        stu_id=id_entry.get()
        con=sql.connect(database="studentdb.sqlite")
        cursor=con.cursor()
        cursor.execute("select bal from students where stu_id=?",(stu_id,))
        row=cursor.fetchone()
        if(row==None):
            messagebox.showwarning("Result","Student does not exist")
        else:
            fee_entry.configure(state="normal")
            fee_entry.delete(0,"end")
            fee_entry.insert(0,row[0])
            fee_entry.configure(state="disabled")
            dep_entry.focus()
    def deposit():
        stu_id=id_entry.get()
        dep_fee=dep_entry.get()
        con=sql.connect(database="studentdb.sqlite")
        cursor=con.cursor()
        cursor.execute("update students set bal=bal-? where stu_id=?",(int(dep_fee),stu_id))
        con.commit()
        con.close()
        messagebox.showinfo('','fee deposited')
        
    frm=Frame(win)
    frm.place(x=0,y=0,relwidth=1,relheight=1)
    
    img_lp = Image.open('d:/fee.jpg')
    tk_imglp = ImageTk.PhotoImage(img_lp.resize((1375,790)))
    lbl_lp = Label(frm,image=tk_imglp)
    lbl_lp.image=tk_imglp
    lbl_lp.pack()
    
    logout_btn=Button(frm,text="Logout",font=('',10),bd=3,command=logout)
    logout_btn.place(relx=.96,rely=0)
    
    home_btn=Button(frm,text="Home",font=('',10),bd=3,command=home)
    home_btn.place(x=0,y=0)
    
    frm1 = Frame(frm)
    frm1.configure(bg='white')
    frm1.place(relx=.44,rely=.13,relwidth=.32,relheight=.37)
    
    stu_id=Label(frm,text="Student id:",font=('',12,'bold'),bg='white')
    stu_id.place(relx=.45,rely=.15)
    
    id_entry=Entry(frm,font=('',12),bd=3)
    id_entry.focus()
    id_entry.place(relx=.6,rely=.15)
    id_entry.bind("<FocusOut>",fetch_fee)
    
    stu_fee=Label(frm,text="Remaining Fee:",font=('',12,'bold'),bg='white')
    stu_fee.place(relx=.45,rely=.25)
    
    fee_entry=Entry(frm,font=('',12),bd=3,state='disabled')
    fee_entry.place(relx=.6,rely=.25)
    
    dep_fee=Label(frm,text="Amount:",font=('',12,'bold'),bg='white')
    dep_fee.place(relx=.45,rely=.35)
    
    dep_entry=Entry(frm,font=('',12),bd=3)
    dep_entry.place(relx=.6,rely=.35)
    
    dep_btn=Button(frm,text="Deposit",font=('',12),bd=3,command=deposit)
    dep_btn.place(relx=.6,rely=.43)

# Add course

def add_course_body():
    def logout():
        frm.destroy()
        main_body()
    def home():
        frm.destroy()
        login_body()
        
    def add_course_db():
        n=name_entry.get()
        n=n.upper()
        f=fee_entry.get()
        if(n==''or f==''):
            messagebox.showinfo('Attention','Incomplete information.')
        else:
            con=sql.connect(database="studentdb.sqlite")
            cursor=con.cursor()
            try:
                cursor.execute("insert into course values(?,?)",(n,f))
                con.commit()
                messagebox.showinfo('','Course Added')
            except Exception as e:
                messagebox.showerror('',str(e))
            con.close()
        
    frm=Frame(win)
    frm.place(x=0,y=0,relwidth=1,relheight=1)
    
    img_lp = Image.open('d:/a_course.jpg')
    tk_imglp = ImageTk.PhotoImage(img_lp.resize((1375,740)))
    lbl_lp = Label(frm,image=tk_imglp)
    lbl_lp.image=tk_imglp
    lbl_lp.pack()
    
    logout_btn=Button(frm,text="Logout",font=('',10),bd=3,command=logout)
    logout_btn.place(relx=.96,rely=0)
    
    home_btn=Button(frm,text="Home",font=('',10),bd=3,command=home)
    home_btn.place(x=0,y=0)
    
    frm1 = Frame(frm)
    frm1.configure(bg='black')
    frm1.place(relx=.28,rely=.28,relwidth=.38,relheight=.52)
    
    course_name=Label(frm,text="Course Name:",font=('',12,'bold'),bg='black',fg='white')
    course_name.place(relx=.3,rely=.4)
    
    name_entry=Entry(frm,font=('',12),bd=3)
    name_entry.focus()
    name_entry.place(relx=.5,rely=.4)
    
    course_fee=Label(frm,text="Course Fee:",font=('',12,'bold'),bg='black',fg='white')
    course_fee.place(relx=.3,rely=.5)
    
    fee_entry=Entry(frm,font=('',12),bd=3)
    fee_entry.focus()
    fee_entry.place(relx=.5,rely=.5)
    
    add_btn=Button(frm,text="Add",font=('',15),bd=3,command=add_course_db)
    add_btn.place(relx=.43,rely=.65)
    
main_body()
win.mainloop()
```
